# 1. Organic Variable Creation

## 1.1 First Organic Variable - The number of coauthors

**Methodology of Variable Extraction**

The number of coauthors variables is generated by counting the total number of coauthors of each publication. In the given data set there is a column called "coauthors". Split each entry by ";" and then count the element in the list generated.

**Justification for Significance of Variable**

The number of coauthors can influence the publication performance in a positive way. Firstly, more coauthors can bring in more expertise and knowledge of both the research area and other relevant academic areas; thus the research team will have more resources and more likely to overcome the challenge during their research. Secondly, more substantial number of coauthors are more likely to have higher cognitive diversity in their team, which can lead to a quicker problem-solving speed. (https://hbr.org/2017/03/teams-solve-problems-faster-when-theyre-more-cognitively-diverse). Lastly, with more coauthors in a group, each member can focus on their research area, sparing less effort to tackle 
problems that they are not familiar with, leading to a higher efficiency.

Indeed, more coauthors can also lead to conflicts more easily, as more people may have different insight in a problem, which will probably lead to lengthy and fierce discussions, hampering the harmony and cooperation within the team. However, those team that cannot solve problems cannot have their research published. Since we are studying the researches that have already been published, we assume that those team in the dataset can maintain well-functioning teamwork.


**Hypothesis of Variable**

The number of coauthors has a positive influence on the citations.


```python
# Split the coauthors column and count the element to get the number of coauthors.
pub['num_coauthors'] = pub.coauthors.apply(lambda x: len(str(x).split(';')))
```

## 1.2 Second Organic Variable - Year dummy variable

**Methodology of Variable Extraction**

The year dummy variables are generated by making the "year" column into category variable, and then use one-hot encoding to get the year dummy variables. The year 2018 excluded as this indicator can be inferred.

**Justification for Significance of Variable**

The year dummy variables should occur naturally as citations should be closely related to the year the paper formally published. Citations of a paper will increase by year as other scholars need time to do their researches, thus the earlier the paper was being published, the more likely the paper is being cited by other papers.

In this research the year is treated as dummy variable instead of producing a new variable-- publish year-- to represent how many years the paper has been published. The main reason behind this is this research is also trying to find out whether some year's research is extremely important. If in a particular year there is a significant breakthrough in the business sector, then that year will have a higher importance score than the other year, even it is later than them.

**Hypothesis of Variable**

The year dummy variables will have a positive influence in the citations. The earlier the year, the coefficient will be larger. 

```python
#use one_hot encoding to get dummy variable and put them into the dataset.
one_hot = pd.get_dummies(pub['year'])
pub = pub.join(one_hot)
```

## 1.3 Third Organic Variable - Topic popularity


**Methodology of Variable Extraction**

The number of other titles has the same nouns in the title is a proxy for topic popularity. The topic popularity variable is generated by firstly removing all the stopwords and punctuations from the "title" column, and then putting the nouns in each title into a list, then comparing the noun list of each pair of publications, and checking the intersection of their title noun list. For each title in each comparison, the popularity adds one if the intersection is not empty. The larger the popularity is, the more popular the topic.

**Justification for Significance of Variable**

The topic popularity is generated because the more popular the research topic is, the more future related research will occur, leading to more citations in this paper. Since the title is the first thing that jump into the eyes and generalizes the topic of the publication, it is natural for researchers to have a closer look into the paper if the title includes some words that are related to their own research.

**Hypothesis of Variable**

The topic popularity variables have a positive influence on the citations.

```python
#remove stopwords and punctuation
# Initialize `tot_words`
tot_words=[]
# Loop through `title` and populate tot_words
for s in pub.title:
    word_list = nltk.word_tokenize(s)
    refined_list = [word for word in word_list if word not in (stopwords.words('english') and string.punctuation)] #remove punctuation and stopwords
    new_title = ' '.join(refined_list)
    new_title = new_title.lower()
    tot_words.append(new_title)
```

```python
# extract noun
def extract_nouns(text):
    # use tokenizer to split text into individual words
    words = nltk.word_tokenize(text)    
    # use tagger to categorize all words
    tags = nltk.pos_tag(words)
    # keep nouns only (nn, nns, nnp, nnps)
    only_nouns = []
    for word, pos in tags: 
        if (pos == 'NN' or pos == 'NNS' or pos == 'NNP' or pos == 'NNPS'):
            only_nouns.append(word.lower())
    return only_nouns  

noun_title = []
for i in tot_words:
    h = extract_nouns(i)
    noun_title.append(h)
```

```python
#popularity calculation
popularity = []
for i in noun_title:
    pop = 0
    for j in noun_title:
        if len(set(i)&set(j))!=0:
            pop = pop + 1
    popularity.append(pop)      
        
pub['topic_popularity'] = popularity  
```

## 1.3 Fourth Organic Variable - Title length

**Methodology of Variable Extraction**

The title length is generated by calculating the number of words in the title.

**Justification for Significance of Variable**

Researches have shown that the length of a title has a negative influence on the citations[1]. As a clear, tightly constructed titles can improve the chances of a paper being discovered by readers and encourage them not to skip it.

**Hypothesis of Variable**

The title length variables have a negative influence on the citations.

```python
#split the title and count the word
pub['title_length'] = pub.title.apply(lambda x:len(str(x).split(' ')))
```

# 2. Inorganic Varaible Creation

In the inorganic variable creation part, two new datasets is used. The "WOS_scraped_df_ver3.pkl" is scraped by Seongmin Lee. This dataset has some useful variables, including the number of reference for each publication,  the abstract of each publication, keywords. ## 2.1 First Inorganic Variable - Number of references

**Methodology of Variable Extraction**

The number of reference variable can be found in the "WOS_scraped_df_ver3.pkl". After merging the two data set on the original dataset's 'title' column and the scraped dataset's 'Title' column, the 'num_of_ref' shows up in the 'pub' dataset. Before the merging, make both 'title' column into lowercase to make sure they can match.

**Justification for Significance of Variable**

The number of references is an important indicator of the quality of the publication. With a sufficient number of reference, the publication itself is more solid as the authors have more backup knowledge. Besides, more reference manifest that the author is one proxy for the amount of endeavor that the author put into the research. Last but not the least, a sufficient number of references reflects that this research topic is interesting and worth studying, and in the future more people will research this topic too, leading to more citations of the publication itself.

## 2.2 Second Inorganic Variable - Journal H index

**Methodology of Variable Extraction**

The journal H index can be found in the "scimagojr 2017.xlsx" in the column H index. Merge the 'pub' and 'hindex[['Title','H index']]' dataframe on 'journal' and 'Title' column. Before doing this step, make sure to transform both columns to lowercase to have a proper match.

**Justification for Significance of Variable**

The journal H index is the least citation of all paper citations in the journal in a given period of time, it can manifest the influence of the journal itself. An influential journal can help the paper published gain more exposure. In the meantime, the journal itself can work as a guarantee of the paper quality, thus promoting the citations of publications.

**Hypothesis of Variable**

The journal H index has a positive influence on the citations.

```python
hindex['Title'] = hindex['Title'].str.lower()
pub['journal'] = pub['journal'].str.lower()
pub_full = pub.merge(hindex[['Title','H index']], 
                     left_on='journal', 
                     right_on = 'Title', 
                     how = 'left')
```

```python
#subset the dataframe to pub_all, which include all the column we need and dump all column we don't need anymore.
list(pub_full)
pub_all = pub_full[['person_id',
                    'coauthors',
                    'title',
                    'journal',
                    2013,
                    2014,
                    2015,
                    2016,
                    2017,
                    2018,                   
                    'citations',
                    'num_coauthors',
                    'topic_popularity',
                    'title_length','num_of_ref',
                    'abstract',
                    'keywords',
                    'H index']]


#dropna to make sure each cells have entry.
pub_all = pub_all.dropna()

#reset index, in the later steps we can do train-test split.
pub_all = pub_all.reset_index(drop = True)
```

## 2.3 Third Inorganic Variable - Abstract Clarity

**Methodology of Variable Extraction**

The abstract clarity shows how clear the abstract is when it explains the method and the result of the paper, it is calculated by counting the number of words that indicating the method and result( including “method”, “result”, “conclusion”,"objective", "discover"). During the counting, should make sure both the target word's plural and singular form are counted.

**Justification for Significance of Variable**

This variable can be useful because a more explicit abstract help other researchers understand the paper more quickly, thus potentially leads to a higher citation. The keyword used here are words that can help readers quickly grasp the results and method of the research, which is the most important things that readers care about when they are reading a research paper.

**Hypothesis of Variable**

The abstract clarity has a positive influence on the citations.

```python
#defind the function to count how many times the target words appear in a string.
def count_word(string, pattern):
    string = string.lower()
    pat = r'(\w*%s\w*)' % pattern  
    length = len(re.findall(pat, string))
    return length

def count_the_word(word_list,string):
    total = 0
    for i in word_list:
        total = total + count_word(string, i)
    return total
```

```python
words = ['result', 'conclusion','method','objective', 'discover']

pub_all['abstract_clarity'] = pub_all['abstract'].apply(lambda x: count_the_word(words, x))
```

The number of keywords counts the total number of keywords for each publication.

**Justification for Significance of Variable**

This variable can be useful because the keywords of a paper demonstrate it's research topic, related theory, application area, research method. The more keywords, the more likely that the paper will show up on the front page of search engines, thus more likely to grasp the attention of researchers, leading to higher citations.

**Hypothesis of Variable**

The number of keywords has a positive influence on the citations.

### 3.1 Detecting Sources of Multicollinearity

the variance inflation factor (VIF) is the ratio of variance in a model with multiple terms, divided by the variance of a model with one term alone. It quantifies the severity of multicollinearity in an ordinary least squares regression analysis.(https://en.wikipedia.org/wiki/Variance_inflation_factor)

The VIF of all variables created is relatively low; thus we will keep all variables and find the feature importance of them in the random forest. 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
for i in range(predictor.shape[1]):
    vif = variance_inflation_factor(predictor.values, i)
    print(predictor.columns[i],':',vif)
```

Since the VIF of number of keyword is 11, higher than the threshold, thus in the regression part, this variables should be excluded.

### 3.2 Feture Selection via Feature Importance

The feature selection part in the random forest is base on the importance of each variable. One way to measure the importance is the Inc node purity. At each split, the algorithm calculates how much this split reduces node impurity (for regression trees, indeed, the difference between RSS before and after the split). This is summed over all splits for that variable, over all trees. The larger the importance, the more important the variables is.

In the "Feature importance in random forest algorithm" graph, we can see the importance of each variable created. The most important variables is "num_of_ref", with the importance of 0.27, followed by "num_coauthors", whose importance is 0.2. The importance of "H index" and "title length" are both higher than 0.1, and the importance of "abstract clarity" is relatively high too, which is 0.08. 

The importance of "topic popularity", year dummy variable and "keywords_num" are relatively low. In the regression model, we should not consider "topic popularity" and "keywords_num" as important variables. However, the year dummy variable is still considered important as they can have large importance if separate importance is added up.

What worth noticing in here is the year 2017 dummy variables, its importance is larger than the year 2015 and year 2016, which can give an insight that in this year some special thing may happen, leading to the papers published in this year have different citation pattern than the other.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#random forest
predictor_list = list(predictor.columns)
citations = np.array(pub_all['citations'])
predictors = np.array(predictor)

train_features, test_features, train_labels, test_labels = train_test_split(predictors, 
                                                                            citations, test_size = 0.25, random_state = 42)                 
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)                
rf.fit(train_features, train_labels)

predictions = rf.predict(test_features)
errors = abs(predictions - test_labels)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'citations')
```

```python
importances = list(rf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(predictors, round(importance, 2)) for predictors, importance in zip(predictor_list, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

feature_importance_full = feature_importances
```

# 4. Predicting Publication Performance - Regression Model and Interpretation

In this part, Non-Negative Binomial regression is used to predict the citations, because the dependent variable is count data, and it is heavily right-skewed. It is also because the citations is over disperse, having way higher variance compared to the mean.

## 4.1 Comparison between regression on all variables and important variables

We build tow non negative binomial regression model in this part, the first model use all variables created, while the second model use the important variable we selected in the variable evaluation part. 

In this part, we split the data into a training set and test set, and train both model in the training set, then test the model in the test set. We compare the MSE, AIC, and devience residual of both model to see which one perform better. 

In the MSE Comparison graph, we can see the MSE of the full model is larger than the MSE of truncated model. And the AIC and the deviance residual of the truncated model are both lower than the full model. Thus we will use this model to do the interpretation part and see how each variable influence the dependent variables.

```python
import statsmodels.formula.api as smf 
import statsmodels.api as sm

df = pub_all[[ 2013,2014,2015,2016,2017,
              'citations','num_coauthors','topic_popularity','title_length','num_of_ref',
              'abstract_clarity', 'keywords_num','H index']]
df.columns = ['year_2013','year_2014','year_2015','year_2016','year_2017','citations','num_coauthors','topic_popularity','title_length','num_of_ref',
              'abstract_clarity', 'keywords_num','H_index']
```

```python
X = ['year_2013','year_2014','year_2015','year_2016','year_2017',
     'num_coauthors','topic_popularity','title_length','num_of_ref',
     'abstract_clarity', 'keywords_num','H_index']

X_important = ['year_2013','year_2014','year_2015','year_2016','year_2017',
               'num_coauthors','title_length','num_of_ref',
               'abstract_clarity', 'H_index']
```

```python
#Running Non Negative Binomial Regression

# Regression using all variables
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error

# Split into training and test sample
train_X, test_X, train_y, test_y = train_test_split(df[X], df['citations'], test_size = 0.20, random_state = 42)

# fit the model onto the training sample
model_nbr = smf.glm(formula = 'citations ~ year_2013 + year_2014 + year_2015+year_2016+year_2017+num_coauthors +topic_popularity+title_length+num_of_ref + abstract_clarity+keywords_num+H_index',
                    data = train_X.join(train_y), 
                    family = sm.families.NegativeBinomial()).fit()
y_hat = model_nbr.predict(test_X)
mse_full =  mean_squared_error(test_y, y_hat)
```

```python
# Regression using important variables
train_X, test_X, train_y, test_y = train_test_split(df[X_important], df['citations'], test_size = 0.20, random_state = 42)

model_nbr2 = smf.glm(formula = 'citations ~ year_2013 + year_2014 + year_2015+year_2016+year_2017+num_coauthors +title_length+num_of_ref + abstract_clarity+H_index',
                     data = train_X.join(train_y), 
                     family = sm.families.NegativeBinomial()).fit()



y_hat_trunc = model_nbr2.predict(test_X) # make the predictions by the model
mse_trunc =  mean_squared_error(test_y, y_hat_trunc)
```

```python
print("MSE for the full model is:",  mse_full)
print("MSE for the truncated model is:",  mse_trunc)

mse = {'Features':["All",'Important'], 'mse':[mse_full,mse_trunc]}
mse = pd.DataFrame(mse)
mse.to_csv("mse.csv")
```

## 4.2 Model interpretation

It can be seen from the truncated model that the p-value of the number of references, H index of a journal and year dummy are all statistically significant, using 95% confidential level. The p-value of title length, number of coauthors are quite small, their marginal effect can also be considered too. In here the "abstract clarity" variable is not significant.

The marginal effect of the number of references is that one unit increase in the number of references will increase the citations by 0.0085 units. 

The marginal effect of H index is that one unit increase in the journal H index will increase the citations by 0.0043 units. This result shows that publishing paper in a more influential journal does help to increase the performance of the publications.

The coefficient of year dummy variables is all positive, which means that compared to the year 2018, the paper published before have higher citations. The marginal effect of year dummy variable compared to the year 2018 is the value of their corresponding coefficient.

The marginal effect of title length is that one unit increase in the title length will decrease the citations by 0.0295 units. This result inspires researchers to make the title more concise.

The marginal effect of the number of coauthors is that one unit increase in the number of coauthors will increase the citations by 0.0692 units. This result shows that more coauthors do help to increase the citations.

The result of abstract clarity variable is the most interesting one, as its coefficient is negative, which means that the word indicating the results and methods of the paper is not important regarding citations. Thus researchers don't need to use too much those words in their abstract.
